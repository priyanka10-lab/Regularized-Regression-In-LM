```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import statsmodels.api as sm
import sklearn.linear_model as lm

data = pd.read_csv("test_sample.csv")

X = data.iloc[:,1:]
Y = data.iloc[:,0]
```

```python
#Lasso
lassoModel = lm.LassoCV(cv=5, random_state=1)
lassoModel.fit(X, Y)

bestlam = lassoModel.alpha_
```

```python
#Fit lasso to the entire sample using best Î».
lassoOut = lm.Lasso(alpha=bestlam)
lassoOut.fit(data.iloc[:,1:] ,data.iloc[:,0])

zero_array_lasso = [lassoOut.coef_ == 0]

results_lasso = np.where(zero_array_lasso)[1]
print(results_lasso)
print(len(results_lasso))
```

```python
#Linear Regression
allPredictors_lm = sm.OLS(Y, sm.add_constant(X)).fit()
print(allPredictors_lm.summary())
```

```python
array_lm = [allPredictors_lm.pvalues > 0.05]

results_lm = np.where(array_lm)[1] - 1 

results_lm = results_lm[1:] 

print(results_lm)
print(len(results_lm))
```

```python

eliminated_by_Lasso = results_lasso
eliminated_by_lm = results_lm

lasso_zeros = ' '.join([str(idx) for idx in eliminated_by_Lasso])
lm_zeros = ' '.join([str(idx) for idx in eliminated_by_lm])
pd.DataFrame([lasso_zeros,lm_zeros], index = ['eliminated_by_Lasso','eliminated_by_lm']).to_csv('answer.csv')
```
